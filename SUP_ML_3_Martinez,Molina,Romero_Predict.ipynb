# ML Preprocessing

```python
del(X_pred["infobase"])
```

```python
columnas_unknown_categoricas = ["prizm_social_one", "area", "refurb_new", "ownrent", "dwlltype", "HHstatin", "dwllsize", "ethnic"]
```

```python
for i in columnas_unknown_categoricas:
    X_pred[i].fillna("Unknown", inplace=True)
```

```python
X_pred["hnd_webcap"].fillna("UNKW", inplace=True)
```

```python

columnas_U_categoricas = ["dualband", "marital", "kid0_2", "kid3_5", "kid6_10", "kid11_15", "kid16_17", "creditcd"]
```

```python
for i in columnas_U_categoricas:
    X_pred[i].fillna("U", inplace=True)
```

```python
mean_eqpdays =X_pred["eqpdays"].mean()
X_pred.loc[(X_pred["eqpdays"]<0), "eqpdays"] = mean_eqpdays
```

```python

nullColumns = ["numbcars", "truck", "rv", "forgntvl"]
```

```python
for i in   nullColumns:
    X_pred[i].fillna(0, inplace=True)
```

```python
NumericOutlayers = ["avg6mou", "avg6qty", "avg6rev", "lor", "income", "adults", "hnd_price", "change_mou", "change_rev", "rev", "mou", "totmrc", "da", "ovrmou", "ovrrev", "vceovr", "datovr", "roam", "phones", "models", "eqpdays", ]
for i in   NumericOutlayers:
    X_pred[i].fillna(-99999, inplace=True)
```

```python
print(f"{len(X_pred.columns[X_pred.isnull().any()])}")
```

```python
 BoolianIdentifiers = {
    "O":1,
    "M":1,
    "Y":1,
    "T":1,
    "WCMB":1,
    "R":0,
    "S":0,
    "U":0,
    "N":0,
    "WC":0,
    "UNKW":-99999,
    "Unknown":-99999
}
```

```python
encodingColumns = ["ownrent", "dwlltype", "kid0_2", "kid3_5", "kid6_10", "kid11_15", "kid16_17", "creditcd", "asl_flag","hnd_webcap"]
```

```python
for i in   encodingColumns:
    X_pred[i].replace(  BoolianIdentifiers, inplace = True)
```

```python

  BoolianIdentifiers_2 = {
    "R":0,
    "N":1,
    "Unknown":-99999
}
```

```python
X_pred["refurb_new"].replace(  BoolianIdentifiers_2, inplace=True)
```

```python

  BoolianIdentifiers_3 = {
    "Y":1,
    "N":0,
    "T":1,
    "U":-99999,
    "Unknown":-99999
}
```

```python
encodingColumns_2 = ["dualband", "new_cell"]
```

```python
for i in   encodingColumns_2:
    X_pred[i].replace(  BoolianIdentifiers_3, inplace = True)
```

```python
replace_dict = {'U': -99999,
                 'M': 1,
                 'S': 0,
                 'B': 0,
                 'A': 0}
```

```python

X_pred['marital'] = X_pred['marital'].replace(replace_dict)
```

```python

for i in   encodingColumns:
    print(X_pred[i].value_counts())
```

```python
X_pred["refurb_new"].value_counts()
```

```python
for i in   encodingColumns_2:
    print(X_pred[i].value_counts())
```

```python

X_pred['marital'].value_counts()
```

```python
X_pred["crclscod"].value_counts().sort_index()
```

```python

categories_to_keep = ['AA', 'A', 'BA', 'CA', 'EA', 'B', 'DA', 'ZA', 'C']
```

```python
X_pred['crclscod'] = X_pred['crclscod'].where(X_pred['crclscod'].isin(categories_to_keep), 'Others')
```

```python
X_pred["crclscod"].value_counts().sort_index()
```

```python
X_pred["dwllsize"].value_counts().sort_index()
```

```python

from sklearn.preprocessing import OrdinalEncoder
```

```python
categorias_dwllsize = X_pred["dwllsize"].value_counts().sort_index().index.tolist()
```

```python
ordinal_encoder_2 = OrdinalEncoder(categories=[categorias_dwllsize])
categorias_dwllsize
```

```python

X_pred['dwllsize'] = ordinal_encoder_2.fit_transform(X_pred[['dwllsize']])
X_pred.loc[X_pred['dwllsize'] == 15, 'dwllsize'] = -99999
```

```python
ethnic_RDMPX = ['R', 'i', 'i', 'P', 'X']
```

```python

X_pred['ethnic'] = X_pred['ethnic'].replace(ethnic_RDMPX, 'R-i-i-P-X')
```

```python
X_pred['ethnic'].value_counts()
```

```python
X_pred["area"].value_counts()
```

```python
X_pred["area"] = X_pred["area"].str.split(' AREA').str[0]
```

```python

X_pred['area'] = X_pred['area'].replace(['NORTH FLORIDA', 'SOUTH FLORIDA'], 'NORTH/SOUTH FLORIDA')
```

```python
X_pred['area'].value_counts()
```

```python

cols_name = ["area", "ethnic", "crclscod"]
cols_category = X_pred[cols_name]
```

```python
for i in cols_category:
    print(X_pred[i].value_counts())
```

```python
for col in cols_category:
    X_pred = pd.get_dummies(X_pred, columns=[col])
```
